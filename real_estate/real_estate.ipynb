# Исследование объявлений о продаже квартир

## Введение

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. 

**Цели**:
- Нужно научиться определять рыночную стоимость объектов недвижимости. 
- Установить параметры. Рассмотреть влияние параметров друг на друга, а так же определим их наиболее вероятные значения при работе на рынке недвижимости. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)
- `cityCenters_nearest` — расстояние до центра города (м) (`city_centers_nearest`)
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
- `first_day_exposition` — дата публикации
- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — общая площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении
- `parks_around3000` - количество парков в радиусе 3000 метров
- `ponds_around3000` - количество водоёмов в радиусе 3000 метров

Но из жизненного опыта и таблицы зависимости столбцов друг от друга, мы знаем, что жилая площадь главным образом зависит от общей площади (`total_area`)и количества комнат(`rooms`). Эти столбцы не имеют пропусков.

Поэтому мы можем заполнить пропуски по категориям. Категории будем определять по количеству комнат. Далее вычислим отношение медианной жилой площади и медианной общей площади. И использую это отношение посчитаем жилую площадь от общей площади для каждого пропущенного значения.

```python
# проверим корректность данных по комнатам
data['rooms'].value_counts() 
```

#### Гипотиза 3 или борьба с неявными дубликатами

**Гипотеза 3** В столбце `locality_name` названия населенных пунктов записаны по разному, например, *Кудрово* и *деревня Кудрово*

```python
data['locality_name'].unique()
```

Из набора названий видно:
- все названия населенных пунктов указаны с большой буквы и последними в строке,
- у городов не указано, что это город,
- в названиях других населенных пунктов много вариантов написания, например е и ё, для посёлков может указываться городской тип в любом месте, а с поселками у жд станции и садовыми товариществами вообще беда.

Конечно, самый простой способ - это оставить только, то что после заглавной буквы. Но тогда все деревни и поселки будут как города. Устроит ли нас это при анализе? Скорее "нет" чем "да". Используем алгоритм посложнее.

```python
# Вхождение -- замена, и обращаем внимание на порядок слов и на последовательность замен

# "пос" --  "посёлок" v
# "кот" "пос" --  "посёлок" 
# "пос" "гор" --  "посёлок городского типа" v v
# "пос" "гор" "имен" -- "посёлок городского типа имени"  v
# "пос" " станц" -- "посёлок при жд станции" v
# "дер"  -- "деревня" v
# "сел" -- "село" v
# "сад" "тов" -- "садовое товарищество" v v

data['locality_name'] = data['locality_name'].replace(regex=
                                                      {
                                                        r'^пос.+станции\s': 'посёлок при жд станции ',
                                                        r'^пос.+имени\s': 'посёлок городского типа имени ',
                                                        r'^садо.+тов.+\s': 'садовое товарищество ',
                                                        r'.*тов.+\s.*садо.+\s': 'садовое товарищество ',                                                       
                                                        r'^городск.*\s.*пос.+\s': 'посёлок городского типа ',
                                                        r'^пос.+\s.*городск.+\sтипа\s': 'посёлок городского типа ', 
                                                        r'поселок': 'посёлок ',
                                                        r'^дер.+\s': 'деревня ',
                                                        # r'^сел.+\s': 'село '
                                                        r'\s+': ' ',
                                                        'деревня Кудрово': 'Кудрово'
                                                      }
                                                     )
# проверка
# data.loc[data['locality_name'].str.find('жд') > -1, 'locality_name'].value_counts()
data['locality_name'].unique()
```

### Задание: Посчитайте среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделите населённые пункты с самой высокой и низкой стоимостью квадратного метра.

```python
top10_locations = data.query('days_exposition > 0') \
    .pivot_table(index='locality_name', values='sq_m_price', aggfunc=['sum', 'count']) \
    .sort_values(by=('count', 'sq_m_price'), ascending=False) \
    .head(10)
# display(top10_locations.columns)

top10_locations.columns = ['sum', 'count']
# display(top10_locations.columns)
#display(top10_locations)

# Конечно, средняя цена средних цен за квадратный метр это не совсем то, что нам нужно, но будет интересно сравнить
top10_locations['avr_avr_sq_m_price'] = top10_locations['sum'] / top10_locations['count']

# округлим и изменим тип данных
top10_locations['avr_avr_sq_m_price'] = np.ceil(pd.to_numeric(top10_locations['avr_avr_sq_m_price'], errors='coerce')).astype('Int64')
# display (top10_locations)

data_top10_locations = data.query('locality_name in @top10_locations.index and days_exposition > 0')
# data_top10_locations.head()

# можно решить через pivot_table как ниже, но хочется испытать цикл и groupby
# top10_locations['avr_sq_m_price'] = 0
for column_value, column_slice in data_top10_locations.groupby('locality_name'):
#    display(column_value, column_slice['last_price'].sum() / column_slice['total_area'].sum())
    top10_locations.loc[top10_locations.index == column_value, 'avr_sq_m_price'] = \
                    column_slice['last_price'].sum() / column_slice['total_area'].sum()


# округлим и изменим тип данных
top10_locations['avr_sq_m_price'] = np.ceil(pd.to_numeric(top10_locations['avr_sq_m_price'], errors='coerce')).astype('Int64')
top10_locations
```

В колонке `avr_sq_m_price` таблицы выше вы можете увидеть среднюю цену за 1 квадртаный метр недвижимости в 10 населенных пунктах где больше всего объявлений. 

**Вывод 1**: Как вы видите самый активный и самый дорогой это Санкт-Петербург и его ближайшие окрестности.
А Выборг из этой выборки имеет самые низкие цены на один квадратный метр нежвижимости.

Справочно, в стобце `avr_avr_sq_m_price`, мы посчитали среднюю средних цен каждой квартиры за один квадртаный метр. Как видите значения в столбцах отличаются в обе стороны. Вероятно эти различия связаны с неоднородностью цены одного квадратного метра по квартирам. 

*Факультативно* сделаем расчеты средних цен для разнокомнатных квартир в этих же 10 населённых пунктах.

```python
data_top10_locations_pivot_table = \
data_top10_locations.pivot_table(index=['locality_name', 'studio', 'rooms']
                                 , values=['last_price', 'total_area']
                                 , aggfunc='sum')
data_top10_locations_pivot_table['avr_sq_m_price'] = data_top10_locations_pivot_table['last_price'] / \
                                                    data_top10_locations_pivot_table['total_area']

# округлим и изменим тип данных
data_top10_locations_pivot_table['avr_sq_m_price'] = \
np.ceil(pd.to_numeric(data_top10_locations_pivot_table['avr_sq_m_price'], errors='coerce')).astype('Int64')
# сделаем так, чтобы было видно все строки
pd.set_option('display.max_rows', None)
data_top10_locations_pivot_table
```

### Задание:  Как стоимость объектов зависит от расстояния до центра города.

```python
# 'city_centers_nearest_km'
# округлим в большу сторону и сделаем целочисленными расстояния до центра
data_top10_locations['city_centers_nearest_km'] = \
np.ceil(pd.to_numeric(data_top10_locations['city_centers_nearest_km'], errors='coerce')).astype('Int64')
```

Расчитаем средние цены одного квадратного метра по типам квартир в Санкт-Петербурге в зависимости от расстояния до центра города.

```python
data_spb_sq_m_price_km = \
(data_top10_locations
                 .query('locality_name == "Санкт-Петербург"')
                 .pivot_table(index=['studio', 'rooms', 'city_centers_nearest_km']
                              , values=['last_price', 'total_area']
                              , aggfunc='sum')
)

data_spb_sq_m_price_km['avr_sq_m_price'] = data_spb_sq_m_price_km['last_price'] / data_spb_sq_m_price_km['total_area']

data_spb_sq_m_price_km
```

Создадим столбце `city_centers_nearest_km` с расстояниями от центра.

```python
# data_spb_sq_m_price_km.index
data_spb_sq_m_price_km['city_centers_nearest_km'] = data_spb_sq_m_price_km.index.get_level_values(2)

# изменим тип данных на целый
data_spb_sq_m_price_km['city_centers_nearest_km'] = pd.to_numeric(data_spb_sq_m_price_km['city_centers_nearest_km']
                                                                  , errors='coerce', downcast='signed')

data_spb_sq_m_price_km
```

```python
data_spb_sq_m_price_km['avr_sq_m_price'].corr(data_spb_sq_m_price_km['city_centers_nearest_km'])
```

**Вывод 1**: Можно утверждать, что при грубой оценке **в Санкт-Петербурге при удалении от центра города цена за одни квадртаный метр падает.**

Сделаем визуализацию и расчеты более детально: будем ориентироваться на количество комнат.

```python
data_spb_sq_m_price_km['rooms'] = data_spb_sq_m_price_km.index.get_level_values(1)
# data_spb_sq_m_price_km.info()
data_spb_sq_m_price_km['studio'] = data_spb_sq_m_price_km.index.get_level_values(0)
data_spb_sq_m_price_km.loc[data_spb_sq_m_price_km['studio'] == False, 'studio'] = 'квартиры'
data_spb_sq_m_price_km.loc[data_spb_sq_m_price_km['studio'] == True, 'studio'] = 'студии'
data_spb_sq_m_price_km
```

```python
for studio in data_spb_sq_m_price_km['studio'].unique():
    for rooms in data_spb_sq_m_price_km.query('studio == @studio')['rooms'].unique():
        df = data_spb_sq_m_price_km.query('studio == @studio and rooms == @rooms')
        cor = df['avr_sq_m_price'].corr(df['city_centers_nearest_km'])
        if pd.isna(cor):
            display ('Для' , studio, 'из', rooms, 'комнат мало данных для расчета корреляция и построения графика')
        else:
            display('Для' , studio, 'из', rooms, 'комнат корреляция средней цены за один квадратный метр к расстоянию от центра в км составляет', cor)
            barplot_name = 'График для '+studio+' из '+ str(rooms) +' комнат.'
            df.plot(x='city_centers_nearest_km',
                  y='avr_sq_m_price',
                  kind='bar',
                  title= barplot_name)
            plt.show()
        
            
```

**Вывод 2**: По данным выше можно более детально посмотреть. Но главное:
- мы можем подтвердить грубую оценку,
- мы можем добавить, что кроме центра так же ценится зона в 6-8 километрах от центра, далее цена снижается. Скорее всего из этого пояса удобно ездить на работу в будни и в центр в выходные.

**Выводы**:
- В базе мало сделок по 2014 году. Лучше эти данные не учитывать.
- Самые продуктивные: цены ниже, сделок больше, огромные суммы всех сделок за год, были в 2017 и 2018 годах.
- 2019 год: цены вверх, количество сделок вниз.
